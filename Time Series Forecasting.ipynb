```python
import pandas as pd
A = pd.read_csv("Desktop/ds/AirPassengers.csv")
```

```python
from warnings import filterwarnings
filterwarnings("ignore")
```

# Convert Date

```python
A.Month = pd.to_datetime(A.Month)
```

# Make Month as index and drop it as a column

```python
A.index = A.Month
A = A.drop(labels=["Month"],axis=1)
```

```python
A.columns = ["Passengers"]
```

# Plot and decompose

```python
import matplotlib.pyplot as plt
plt.plot(A)
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
Q = seasonal_decompose(A,model='multiplicative')
```

```python
plt.figure(figsize=(15,10))

plt.subplot(2,2,1)
Q.observed.plot()

plt.subplot(2,2,2)
Q.trend.plot()

plt.subplot(2,2,3)
Q.seasonal.plot()

plt.subplot(2,2,4)
Q.resid.plot()
```

# Rolling mean and Rolling Stdev

```python
RM_Original = A.rolling(window=12).mean()
RST_Original = A.rolling(window=12).std()
```

```python
plt.plot(A,c="blue")
plt.plot(RM_Original,c="red")
plt.plot(RST_Original,c="yellow")
plt.title("Non Stationary")
```

# Conversion attempt 1
Convert TS --> Log

```python
import numpy as np

log_A = np.log(A)
RM_LOG_A = log_A.rolling(window=12).mean()
RST_LOG_A = log_A.rolling(window=12).std()

plt.plot(log_A,c="pink")
plt.plot(RM_LOG_A,c="red")
#plt.plot(RST_LOG_A,c="brown")
plt.title("Non Stationary")
```

# Conversion attempt 2:

```python
log_A = np.log(A)
RM_LOG_A = log_A.rolling(window=12).mean()
```
